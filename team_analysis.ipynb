```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
plt.rcParams['figure.figsize'] = (30,15)
```

```python
df1 = pd.read_csv('astros batting data 2017.csv')
df1
```

```python
plt.plot_date(df1['Date'], df1['BA'])
xticks([3, 27, 56, 83, 107, 135, 162], ['April', 'May', 'June', 'July', 'August', 'September', 'October'])
plt.title('Date vs. Batting Average (BA)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Batting Average (BA)', fontsize = 20)
```

```python
print(df1['BA'].mean())
```

## An obvious increase in the Batting Average occurs around the same time as the sign stealing campaign began. This is one initial observation that could have a serious foundation in the logic of the cheating scandal.

```python
plt.plot_date(df1['Date'], df1['RBI'], '-')
xticks([3, 27, 56, 83, 107, 135, 162], ['April', 'May', 'June', 'July', 'August', 'September', 'October'])
plt.title('Date vs. Runs Batted In (RBI)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Runs Batted In (RBI)', fontsize = 20)
```

```python
print(df1['RBI'].mean())
```

## RBIs seemed to increase in the middle of the season but also died off a bit as the playoffs approached.

```python
plt.plot_date(df1['Date'], df1['H'], '-')
xticks([3, 27, 56, 83, 107, 135, 162], ['April', 'May', 'June', 'July', 'August', 'September', 'October'])
plt.title('Date vs. Hits (H)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Hits (H)', fontsize = 20)
```

```python
print(df1['H'].mean())
```

## The number of hits increased around the May 21st scandal beginning and remained relatively consistent for the remainder of the season.

```python
plt.plot_date(df1['Date'], df1['OBP'])
xticks([3, 27, 56, 83, 107, 135, 162], ['April', 'May', 'June', 'July', 'August', 'September', 'October'])
plt.title('Date vs. On Base Percentage (OBP)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('On Base Percentage (OBP)', fontsize = 20)
```

```python
print(df1['OBP'].mean())
```

## Seeing as how the cheating began around May 21st, it is interesting how the rest of the season is relatively consistent when compared to the beginning of the season. Could this be a product of cheating or just finding more consistency with a team throughout a season?

```python
plt.plot_date(df1['Date'], df1['AB'], '-')
xticks([3, 27, 56, 83, 107, 135, 162], ['April', 'May', 'June', 'July', 'August', 'September', 'October'])
plt.title('Date vs. At-Bats (AB)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('At-Bats (AB)', fontsize = 20)
```

```python
print(df1['AB'].mean())
```
