```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
plt.rcParams['figure.figsize'] = (30,15)
```

```python
df1 = pd.read_csv('redsox_batting_data_2017.csv')
df1
```

```python
plt.plot_date(df1['Date'], df1['BA'], '-')
df1.BA.rolling(10).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Batting Average (BA)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Batting Average (BA)', fontsize = 20)
```

```python
print(df1['BA'].mean())
```

## The Astros had an overall upwards trending progression throughout the season whereas the Red Sox fell throughout the season. The Astros' average was 0.277 and the Red Sox's was 0.266.

```python
plt.plot_date(df1['Date'], df1['RBI'], '-')
df1.RBI.rolling(6).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Runs Batted In (RBI)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Runs Batted In (RBI)', fontsize = 20)

```

```python
print(df1['RBI'].mean())
```

## The Astros' average RBI value was 5.272 whereas the Red Sox's was 4.537. The Astros' cheating timeline shows a consistently higher RBI statistic when compared to the Red Sox.

```python
plt.plot_date(df1['Date'], df1['H'], '-')
df1.H.rolling(5).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Hits (H)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Hits (H)', fontsize = 20)
```

```python
print(df1['H'].mean())
```

## The Astros averaged 9.759 hits per game whereas the Red Sox averaged 9.019. The Astros again were more consitent during their scandal period.

```python
plt.plot_date(df1['Date'], df1['OBP'],)
df1.OBP.rolling(5).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. On Base Percentage (OBP)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('On Base Percentage (OBP)', fontsize = 20)
```

```python
print(df1['OBP'].mean())
```

## This is an interesting result for the Red Sox and their on-base percentage. Where the Astros averaged a 0.343 and plateaued in the beginning to mid-April, the Red Sox had an oscillatory pattern throughout the entire season.
