```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
plt.rcParams['figure.figsize'] = (30,15)
```

```python
df1 = pd.read_csv('nationals_batting_data_2017.csv')
df1
```

```python
plt.plot_date(df1['Date'], df1['BA'], '-')
df1.BA.rolling(10).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Batting Average (BA) - (WA Nationals)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Batting Average (BA)', fontsize = 20)
```

```python
print(df1['BA'].mean())
```

## When compared to the Astros BA, the Nationals had an opposite trend throughout time. The mean batting average for the Nationals is only 0.001 less than the Astros'.

```python
plt.plot_date(df1['Date'], df1['RBI'], '-')
df1.RBI.rolling(6).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Runs Batted In (RBI)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Runs Batted In (RBI) - (WA Nationals)', fontsize = 20)

```

```python
print(df1['RBI'].mean())
```

## There is a moderate difference when comparing the data here. The Astros averaged 5.27 RBIs per game where the Nationals only averaged 4.91. The Astros had a consistently higher RBI value per game during the cheating scandal.

```python
plt.plot_date(df1['Date'], df1['H'], '-')
df1.H.rolling(5).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Hits (H) - (WA Nationals)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Hits (H)', fontsize = 20)
```

```python
print(df1['H'].mean())
```

## The Astros averaged 9.76 hits per game where the Nationals only averaged 9.11 and again, the Astros' values here were consistently higher than the Nationals.

```python
plt.plot_date(df1['Date'], df1['OBP'],)
df1.OBP.rolling(5).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. On Base Percentage (OBP) - (WA Nationals)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('On Base Percentage (OBP)', fontsize = 20)
```

```python
print(df1['OBP'].mean())
```

## Again, the trends are reversed for this statistic. The Astros started low (~0.24 OBP) and by mid-April, plateaued for the remainder of the season. The Astros averaged a 0.343 OBP where the Nationals averaged a 0.345. This is the only statistic that the Nationals have beat out the Astros.

```python
plt.plot_date(df1['Date'], df1['AB'], '-')
df1.AB.rolling(5).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. At-Bats (AB) - (WA Nationals)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('At-Bats (AB)', fontsize = 20)
```

```python
print(df1['AB'].mean())
```

## The Nationals and Astros saw nearly the same number of at-bats with the Nationals averaging 34.28 and the Astros averaging 34.64 per game. There is not a substantial difference in these plots.
