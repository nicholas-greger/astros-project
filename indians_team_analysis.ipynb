```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
plt.rcParams['figure.figsize'] = (30,15)
```

```python
df1 = pd.read_csv('indians_batting_data_2017.csv')
df1
```

```python
plt.plot_date(df1['Date'], df1['BA'], '-')
df1.BA.rolling(10).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Batting Average (BA) - (Cleveland Indians)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Batting Average (BA)', fontsize = 20)
```

```python
print(df1['BA'].mean())
```

## The Indians, minus an incredible drop in early April, had a relatively consistent batting average for the 2017 season. The Red Sox and Nationals both had a negative trending BA but the Indians maintained a 0.254 average against the Astros' 0.277.

```python
plt.plot_date(df1['Date'], df1['RBI'], '-')
df1.RBI.rolling(6).mean().plot()
plt.axvline(x=50, color = 'red',ls ='--')
plt.axvline(x=152, color = 'red',ls ='--')
xticks([3, 27,50, 56, 83, 107, 135, 152, 162], ['April', 'May', 'Cheating Begins', 'June', 'July', 'August', 'September','Cheating Ends', 'October'])
plt.title('Date vs. Runs Batted In (RBI)', fontsize = 30)
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Runs Batted In (RBI)', fontsize = 20)

```

```python
print(df1['RBI'].mean())
```

## The Indians had a 4.815 RBI average value against the Astros' 5.272 average. The Astros maintain a consistently higher average during their cheating scandal.
